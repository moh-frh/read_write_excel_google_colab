```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from datetime import datetime
import math

from google.colab import files

final = pd.read_excel('rdo.xlsx')
```

```python
def get_type_affaire(type_aff):
  if type_aff.lower() == "qls":
    return 0

  if type_aff.lower() == "ext":
    return 1

  return type_aff

def get_rd_id(rd):
  if rd == 'RDA':
    return str("1")
  if rd == 'RDO':
    return str("2")
  if rd == 'RDC':
    return str("3")
  if rd == 'RDE':
    return str("4")

  return str('******')

  def get_convention(conv):
    if conv == 'RCN':
      return 0
    return 1
  
def get_type_affaire(affaire, energie):
  if affaire.lower().find("er") != -1 and energie == 'elec':
    return str("1")

  if affaire.lower().find("dp") != -1 and energie == 'gaz':
    return str(1)

  if affaire.lower().find("qls") != -1:
    return str(2)

  if affaire.lower().find("ext") != -1:
    return str(3)

  if affaire.lower().find("ai") != -1:
    return str(4)    
```

```python


final['Region de Distribution'] = final.apply( (lambda row : get_rd_id(row[0])), axis=1)
final['Direction de Distribution'] = final.apply( (lambda row : str(row[1]).lower()), axis=1)
final['Convention'] = final.apply( (lambda row : "pas de convention" if str(row[2]).lower() == "nan" else row[2].lower()), axis=1 )

final['Type Affaire'] = final.apply( (lambda row : get_type_affaire(str(row[5]), "gaz")), axis=1)

final['Commune'] = final.apply( (lambda row : str(row[4]).lower()), axis=1)
# final['Type Affaire'] = final.apply( (lambda row : 0 if str(row[5]).lower() == "qls" else (1 if str(row[5]).lower() == "ext" else 99)), axis=1)

final['Intitulé de l\'affaire ZO/Français'] = final.apply( (lambda row : str(row[6]).lower()), axis=1)

final['date début d\'état'] = final.apply( (lambda row : "0" if str(row[9]) == "nan" else str(row[9]) ), axis=1)

final['LRT (km)'] = final.apply( (lambda row : 0 if str(row[10]).lower() == "nan" else row[10] ), axis=1)
final['LRD (km)'] = final.apply( (lambda row : 0 if str(row[11]).lower() == "nan" else row[11] ), axis=1)
final['Brts (un)'] = final.apply( (lambda row : 0 if str(row[12]).lower() == "nan" else row[12] ), axis=1)
final['Poste (un)'] = final.apply( (lambda row : 0 if str(row[13]).lower() == "nan" else row[13] ), axis=1)

final['Montant prévisionnel DA'] = final.apply( (lambda row : 0 if str(row[14]).lower() == "nan" else row[14] ), axis=1)
final['Montant engagé DA'] = final.apply( (lambda row : 0 if str(row[15]).lower() == "nan" else row[15] ), axis=1)
final['Montant déponsé DA'] = final.apply( (lambda row : 0 if str(row[16]).lower() == "nan" else row[16] ), axis=1)

final['Affaires annulées'] = final.apply( (lambda row : "true" if str(row[17]).lower() == "oui" else "false" ), axis=1)
final['Motif d\'annulation'] = final.apply( (lambda row : "vide" if str(row[18]) == "nan" else str(row[18]) ), axis=1)
```
